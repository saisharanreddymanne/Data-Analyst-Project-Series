```python
# now let's change th Budget and Gross Columns too

rd['budget'] = rd['budget'].apply(lambda x: f"{int(x)}$" if not pd.isna(x) else x)
```

```python
rd['gross'] = rd['gross'].apply(lambda y: f"{int(y)}$" if not pd.isna(y) else y)
```

```python
rd.head()
```

`Now Budget and Gross values have update from float to int and added "$" for better visualization. But still I want to trim the vale, Let's convert 19000000$ to 19M$`

```python
def convert_val(val):
  try:
    val = str(val).replace("$", "").strip()

    num = float(val)


    if num >= 1_000_000:
      return f"{int(num/1_000_000)}M$"
    elif num >= 1_000:
      return f"{int(num/1_000)}K$"
    else:
      return f"{int(num)}$"
  except ValueError:
    return val

rd['budget'] = rd['budget'].apply(convert_val)

rd.head()
```

```python
def convert_gross(val1):
  try:
    val1 = str(val1).replace("$", "").strip()

    num1 = float(val1)


    if num1 >= 1_000_000:
      return f"{int(num1/1_000_000)}M$"
    elif num1 >= 1_000:
      return f"{int(num1/1_000)}K$"
    else:
      return f"{int(num1)}$"
  except ValueError:
    return val1

rd['gross'] = rd['gross'].apply(convert_gross)

rd.head()
```

Now let's order the table by Gross to get descending order  of the ,pvies list with gross value

```python
rd=rd.sort_values(by=['gross'], ascending= False, na_position='last')
rd
```
